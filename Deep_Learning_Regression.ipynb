# Understanding why deep learning (neural network) works


## Mathematical formulation of neural network

Let $\phi (w_i x + b_i)$ be the basis function where all $b_i$ are constant. Then $c_i$ be the coefficients of linear regression. 

\begin{equation}
     y = \sum_{i = 0}^{n}c_i \phi(w_i x - b_i)
\end{equation}



### Imports

```
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
```

### Understanding Universal approximation theorem 

```
x = np.linspace(start = -10, stop = +10, num = 200 + 1, endpoint = True)
data = pd.DataFrame(dict(x = x))
```

Applying sigmoid function

\begin{equation}
    \sigma(x) = \frac{1}{1 + e ^ {-x}}
\end{equation}

Applying scaled and shifted sigmoid

\begin{equation}
\sigma(x; w, b) = \frac{1}{1 + e^{-(wx + b)}}
\end{equation}

```
for idx in range(20 + 1):
    data.eval('b_'+str(idx)+' = 1/(1 + exp(-(x + ('+str(idx - 10)+'))))', inplace = True)
    
for idx in range(40 + 1):    
    data.eval('w_'+str(idx)+' = 1/(1 + exp(-(x * ('+str(idx - 20)+'))))', inplace = True)

display(data[['b_'+str(i) for i in range(5)]].head())
```

## Effect of bias ($b_i$) on sigmoid

```
ax_ = None
for idx in range(20 + 1):
    ax_ = data.plot(x = 'x', y = 'b_'+str(idx), ax = ax_, label = 'b = '+str(idx - 10))
```

## Effect of weights on sigmoid

```
display(data[['w_'+str(i) for i in range(5)]].head())
```

```
ax_ = None
for idx in range(0, 40 + 1, 2):
    ax_ = data.plot(x = 'x', y = 'w_'+str(idx), ax = ax_, label = 'w = '+str(idx - 20))
```
